```python
# Asignar `NaN` en estadísticas de porteros para jugadores que no son "GK" solo después de la agrupación
def assign_to_non_gk(df, keepers_columns, value=0):
    df.loc[~df['Pos'].str.contains('GK', na=False), keepers_columns] = value

# Configurar las columnas de cálculo de porcentaje y promedios ponderados
def define_calculation_params():
    additional_calculations = {
        'shooting_SoT_pct': ('shooting_SoT', 'shooting_Sh'),
        'passing_Cmp_pct': ('passing_Cmp', 'passing_Att'),
        'passing_Cmp_pct_short': ('passing_Cmp_short', 'passing_Att_short'),
        'passing_Cmp_pct_mid': ('passing_Cmp_mid', 'passing_Att_mid'),
        'passing_Cmp_pct_long': ('passing_Cmp_long', 'passing_Att_long'),
        'defense_Tkl_pct': ('defense_Tkl_OonO', 'defense_Att'),  # Corrección en el nombre
        'possession_Succ_pct': ('possession_Succ', 'possession_Att'),
        'misc_Won_pct': ('misc_Won', 'misc_Won', 'misc_Lost'),
        'keepers_Save_pct': ('keepers_Saves', 'keepers_SoTA'),
        'keepers_CS_pct': ('keepers_CS', 'stats_MP'),
        'keepers_Save_pct_PK': ('keepers_PKsv', 'keepers_PKatt', 'keepers_PKm'),
        'keepersadv_Cmp_pct': ('keepersadv_Cmp', 'keepersadv_Att'),
        'keepersadv_Stp_pct': ('keepersadv_Stp', 'keepersadv_Opp'),
        'possession_Tkld_pct': ('possession_Succ', 'possession_Tkld')
    }

    weighted_avg_columns = {
        'playingtime_PPM': 'stats_MP',
        'keepersadv_Launch_pct': 'stats_MP',
        'keepersadv_AvgLen': 'stats_MP',
        'keepersadv_Launch_pct_goalkick': 'stats_MP',
        'keepersadv_AvgLen_goalkick': 'stats_MP',
        'keepersadv_AvgDist': 'stats_MP',
        'playingtime_Mn_over_MP': 'stats_MP',         
        'playingtime_Mn_over_Start': 'playingtime_Starts',  
        'playingtime_Mn_over_Sub': 'playingtime_Subs'       
    }
    return additional_calculations, weighted_avg_columns

# # Agrupar el DataFrame y realizar sumas básicas y selecciones
# def aggregate_data(df, fbref_info_columns, additional_calculations):
#     fbref_stats_columns = [col for col in df.columns if col not in fbref_info_columns]
    
#     agg_dict = {
#         **{col: 'sum' for col in fbref_stats_columns if col not in additional_calculations},
#         'Player': 'first',
#         'Nation': 'first',
#         'Age': 'first',
#         'Born': 'first',
#         'liga_pais': 'first',
#         'Squad': lambda x: ','.join(x.unique()),
#         'Squad_id': lambda x: ','.join(map(str, x.unique())),
#         'Pos': lambda x: ','.join(sorted(set(pos.strip() for sublist in x.dropna().str.split(',') for pos in sublist))),
#         'Competition': lambda x: ','.join(x.unique())
#     }
#     return df.groupby(['Player_id', 'temporada']).agg(agg_dict).reset_index()

def aggregate_data(df, fbref_info_columns, additional_calculations):
    fbref_stats_columns_sum = [col for col in df.columns if col not in fbref_info_columns and col not in additional_calculations]
    
    df_to_dic = df[fbref_info_columns]
    df_to_sum = df[['Player_id', 'temporada']+ fbref_stats_columns_sum]
    
    # Crear el diccionario de agregación
    agg_dict = {
        'Player': 'first',
        'Nation': 'first',
        'Age': 'first',
        'Born': 'first',
        'liga_pais': 'first',
        'Squad': lambda x: ','.join(x.unique()),
        'Squad_id': lambda x: ','.join(map(str, x.unique())),
        'Pos': lambda x: ','.join(sorted(set(pos.strip() for sublist in x.dropna().str.split(',') for pos in sublist))),
        'Competition': lambda x: ','.join(x.unique())
    }

    # Aplicar groupby con el diccionario de agregación
    result_text = df_to_dic.groupby(['Player_id', 'temporada']).agg(agg_dict).reset_index()
    result_sum = df_to_sum.groupby(['Player_id', 'temporada']).sum(min_count=1).reset_index()

    result = pd.merge(
        result_text,
        result_sum,
        on=['Player_id', 'temporada'], 
        how='left', 
        indicator=False
    )
    
    return result



# Calcular promedios ponderados
def calculate_weighted_averages(df, original_df, weighted_avg_columns):
    for col, weight_col in weighted_avg_columns.items():
        df[col] = round((
            (original_df[col] * original_df[weight_col]).groupby([original_df['Player_id'], original_df['temporada']]).sum(min_count=1) /
            original_df[weight_col].groupby([original_df['Player_id'], original_df['temporada']]).sum(min_count=1)
        ).reset_index(drop=True), 1)

# Calcular porcentajes adicionales
def calculate_additional_percentages(df, additional_calculations):
    for new_col, (num_col, *den_cols) in additional_calculations.items():
        if len(den_cols) == 1:
            df[new_col] = np.where(
                df[den_cols[0]] == 0,
                np.nan,
                round(100 * df[num_col] / df[den_cols[0]], 1)
            )
        else:
            denominator = df[den_cols[0]] + df[den_cols[1]] if new_col == 'misc_Won_pct' else df[den_cols[0]] - df[den_cols[1]]
            df[new_col] = np.where(
                denominator == 0,
                np.nan,
                round(100 * df[num_col] / denominator, 1)
            )

# Procesar el DataFrame completo con todos los pasos necesarios
def process_df(df, nan_keepers=False):
    fbref_info_columns = ['Player', 'Player_id', 'Squad', 'Squad_id', 'Nation', 'Pos', 'Competition', 'Age', 
                          'Born', 'temporada', 'liga_pais']
    
    additional_calculations, weighted_avg_columns = define_calculation_params()
    
    # Agrupar y agregar datos básicos
    df_aggregated = aggregate_data(df, fbref_info_columns, additional_calculations)
    
    # Calcular promedios ponderados
    calculate_weighted_averages(df_aggregated, df, weighted_avg_columns)
    
    # Calcular porcentajes adicionales
    calculate_additional_percentages(df_aggregated, additional_calculations)
    
    # Restaurar NaN o 0 en estadísticas de porteros para jugadores que no son "GK"
    #Se decide rellenar con ceros porque los jugadores de campo no realizan ese tipo de acciones, no es que faltan
    keepers_columns = [col for col in df.columns if col.startswith('keepers')]
    if nan_keepers:
        assign_to_non_gk(df_aggregated, keepers_columns, value=np.nan)
    else:
        assign_to_non_gk(df_aggregated, keepers_columns)
    
    return df_aggregated
```

##### Creando mas _over90s

```python
def create_per90_stats(df, stats_columns):
    """
    Crea nuevas columnas normalizadas por 90 minutos para las estadísticas que no contengan '90' o '_pct'.
    
    Args:
        df (pandas.DataFrame): DataFrame con las estadísticas
        stats_columns (list): Lista de columnas de estadísticas
    
    Returns:
        pandas.DataFrame: DataFrame con las nuevas columnas agregadas
    """
    import pandas as pd
    import numpy as np
    
    # Crear una copia del DataFrame original
    df_original = df.copy()

    excluir = ['playingtime_PPM', 'gca_GCA', 'gca_SCA', 'playingtime_Min']
    
    # Filtrar columnas que no contengan '90' o '_pct'
    cols_to_normalize = [col for col in stats_columns
                         if '90' not in col and '_pct' not in col
                         and not col.startswith('stats_')
                         and col not in excluir
                         and f"{col}_over90" not in stats_columns]
    
    # Verificar que no vayamos a crear columnas duplicadas
    new_column_names = [f"{col}_over90" for col in cols_to_normalize]
    duplicate_new_cols = set(new_column_names) & set(df_original.columns)
    if duplicate_new_cols:
        raise ValueError(f"Las siguientes columnas '_over90' ya existen en el DataFrame: {duplicate_new_cols}")
    
    
    # Crear un diccionario para almacenar las nuevas columnas
    new_columns = {}
    
    # Calcular todas las nuevas columnas de una vez
    for col in cols_to_normalize:
        new_col_name = f"{col}_overPT90"
        new_columns[new_col_name] = df_original[col] / df_original['playingtime_90s']
        
    # Crear un DataFrame con las nuevas columnas
    df_new_cols = pd.DataFrame(new_columns, index=df_original.index)
    
    # Reemplazar infinitos por NaN
    df_new_cols = df_new_cols.replace([np.inf, -np.inf], np.nan)
    
    # Concatenar el DataFrame original con las nuevas columnas
    return pd.concat([df_original, df_new_cols], axis=1)
```

##### Long to Wide de la Copa Libertadores

```python

pd.options.mode.chained_assignment = None  # Suppress SettingWithCopyWarning

# Paso 1: Crear DataFrames locales e internacionales y usar .copy() para evitar SettingWithCopyWarning
df_local = df_fbref[df_fbref['Competition'] != 'Copa Libertadores']
df_libertadores = df_fbref[df_fbref['Competition'] == 'Copa Libertadores']

# Paso 2: Quitar los primeros 3 caracteres de 'Squad' en `df_libertadores`
df_libertadores['Squad'] = df_libertadores['Squad'].str[3:]

# Paso 3: Aplicar `process_df` a ambos DataFrames
df_local_agg = process_df(df_local)
df_libertadores_agg = process_df(df_libertadores)

# Renombrar columnas en `df_libertadores_agg` con el prefijo 'Internacional_'
fbref_info_columns = ['Player', 'Player_id', 'Squad', 'Squad_id', 'Nation', 'Pos', 'Competition', 'Age', 
                      'Born', 'temporada', 'liga_pais']

cols_to_rename = [col for col in df_libertadores_agg.columns if col not in fbref_info_columns]

df_libertadores_agg = df_libertadores_agg.rename(columns={col: f"Lib_{col}" for col in cols_to_rename})


# Paso 4: Merge de `df_local` y `df_libertadores` en `Player_id` y `temporada`
df_fbref_agg = pd.merge(
    df_local_agg,
    df_libertadores_agg,
    on=['Player_id', 'temporada'],
    how='left',
    suffixes=('', '_int')
)

columnas_lib = df_libertadores_agg.select_dtypes(include=['float64', 'int64']).columns[
    (df_libertadores_agg.select_dtypes(include=['float64', 'int64']).columns.str.startswith('Lib')) & 
    (~df_libertadores_agg.select_dtypes(include=['float64', 'int64']).columns.str.contains('_pct'))
]

# Se rellenan los datos vacios con ceros porque se asume que no jugaron la copa
df_fbref_agg[columnas_lib] = df_fbref_agg[columnas_lib].fillna(0)

# Paso 5: Combinar `Squad` y `Squad_id` de ambos conjuntos excluyendo NaN y eliminando duplicados dentro de sets
def combine_unique_clubs(row, col):
    clubs_local = set(map(str.strip, str(row[col]).split(','))) if pd.notna(row[col]) else set()
    clubs_intl = set(map(str.strip, str(row[f"{col}_int"]).split(','))) if pd.notna(row[f"{col}_int"]) else set()
    combined_clubs = sorted(clubs_local.union(clubs_intl))
    return ','.join(combined_clubs)

for col in ['Squad', 'Squad_id']:
    df_fbref_agg[col] = df_fbref_agg.apply(lambda row: combine_unique_clubs(row, col), axis=1)

# Eliminar columnas duplicadas no necesarias después del merge
df_fbref_agg = df_fbref_agg.drop(columns=[f"{col}_int" for col in [
    'Player', 'Squad', 'Squad_id', 'Nation', 'Pos', 'Age', 'Born', 'liga_pais']])

fbref_agg_info_columns = fbref_info_columns + ['Competition_int']

fbref_agg_stats_columns = [col for col in df_fbref_agg.columns if col not in fbref_agg_info_columns]

df_fbref_agg = df_fbref_agg[fbref_agg_info_columns + fbref_agg_stats_columns]
```
